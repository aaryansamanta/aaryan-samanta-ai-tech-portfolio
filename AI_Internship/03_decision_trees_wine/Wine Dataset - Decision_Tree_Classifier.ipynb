```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Wine dataset
wine = load_wine()

# 2. Convert to DataFrame
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df['target'] = wine.target
df['target_name'] = df['target'].apply(lambda x: wine.target_names[x])

# ðŸ‘‰ Preview the dataset (5 first rows)
print("ðŸ“Š Wine Dataset Preview:")
print(df.head())

# 3. Prepare features and labels

X = wine.data
y = wine.target

# 4. Split the dataset (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Initialize and train the Decision Tree model (use parameters: max_depth=4, random_state=42)

#hint: search for 'Create a DecisionTreeClassifier instance'
model = DecisionTreeClassifier(max_depth=4, random_state = 42)

#Predict the response for test dataset

#hint: search for Train the model for DecisionTreeClassifier
model = model.fit(X_train,y_train)

# 6. Make predictions
y_pred = model.predict(X_test)

# 7. Evaluate model performance using accuracy_score function
accuracy = accuracy_score(y_test, y_pred)
print(f"\nâœ… Model Accuracy: {accuracy * 100:.2f}%")
print("\nðŸ“„ Classification Report:")
print(confusion_matrix(y_test, y_pred))

# Confusion matrix


# 8. Visualize the decision tree
plt.figure(figsize=(12, 6))
plot_tree(model, feature_names=wine.feature_names, class_names=wine.target_names, filled=True, fontsize=8)
plt.title("ðŸŒ³ Decision Tree Visualization")
plt.show()

# 9. Predict the class of a new wine sample
# Using the first sample as a test

```
