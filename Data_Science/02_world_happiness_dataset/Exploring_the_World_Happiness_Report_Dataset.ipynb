```python
from google.colab import files
uploaded = files.upload()
import pandas as pd

# Load the World Happiness Report dataset
url = "https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2021"
data = pd.read_csv("world-happiness-report-2021.csv")

# Display the first 5 rows
print(data.head())

# Print the dataset shape
print(f"The dataset has {data.shape[0]} rows and {data.shape[1]} columns.")

# Show a summary of the dataset
print(data.info())

# Show statistical details of numerical columns
print(data.describe())
```

The World Happiness Report 2021 dataset has 149 rows and 20 columns, for each of the 149 countries and their respective indicators measuring happiness. Using the function .info() , I found out that the dataset has 18 numeric columns and 2 categorical columns. The 2 categorical columns, namely “Country” and “year,” have unique values for every row so there are no missing values in this dataset. Therefore, there are no gaps to fill in the data, and the data is complete and intact. Using the function .describe() provided the key statistics, mean, min, max, and quartile of each of the numeric columns and provided concrete insight into how countries generally performed based on one of the scores.
I found it really interesting that Finland is listed as the happiest country with a ladder score of 7.842 while the ladder score for the lowest score was 2.523, which indicates a huge gap in overall happiness globally. Also, healthy life expectancy varied in this data from about 48 years to almost 77 years, showing a level of variation between nations and places in terms of healthy healthcare and living conditions. Another interesting surprise I found from my exploration of the data is that the “Generosity” column had negative values, indicating that generosity was being evaluated; either by wealth or perception as a measure of generosity rather than strictly a positive number.
