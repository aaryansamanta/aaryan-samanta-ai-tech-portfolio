```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Math': [85, 92, 78, 88, 92, 80, 70, 91, 85, 78],
    'Science': [80, 85, 75, 90, 93, 78, 82, 88, 79, 84],
    'English': [88, 92, 85, 90, 89, 87, 81, 94, 86, 90]
}
df = pd.DataFrame(data)

stats_summary = {}
for subject in ['Math', 'Science', 'English']:
    stats_summary[subject] = {
        'Mean': np.mean(df[subject]),
        'Median': np.median(df[subject]),
        'Mode': stats.mode(df[subject], keepdims=True).mode[0],
        'Variance': np.var(df[subject], ddof=1)
    }

stats_df = pd.DataFrame(stats_summary)
print("Descriptive Statistics:\n")
print(stats_df)

plt.figure(figsize=(12,4))
for i, subject in enumerate(['Math','Science','English'],1):
    plt.subplot(1,3,i)
    plt.hist(df[subject], bins=5, alpha=0.7, edgecolor="black")
    plt.title(f"{subject} Scores Distribution")
plt.tight_layout()
plt.show()

df[['Math','Science','English']].plot(kind='box', figsize=(8,5))
plt.title("Score Distributions (Boxplot)")
plt.show()

total_cards = 52
red_cards = 26
black_cards = 26
kings = 4

p_red = red_cards / total_cards
p_black = black_cards / total_cards
p_king = kings / total_cards

p_two_red = (red_cards/total_cards) * ((red_cards-1)/(total_cards-1))

print("\nProbability Results:")
print(f"P(Red card) = {p_red:.4f}")
print(f"P(Black card) = {p_black:.4f}")
print(f"P(King) = {p_king:.4f}")
print(f"P(Two red cards without replacement) = {p_two_red:.4f}")

corr_matrix = df[['Math','Science','English']].corr()
print("\nCorrelation Matrix:")
print(corr_matrix)

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```
